```python
def extract_tweets(id_, max_results):
    result = []
    
    #extract tweets and add to result
    user_tweets = client.get_users_tweets(id = id_, max_results = max_results).data
    
    for tweet in user_tweets:
        tweet_list = []
        tweet_list.append(client.get_user(id = id_).data.id)
        tweet_list.append(client.get_user(id = id_).data.name)
        tweet_list.append(tweet.text)
        result.append(tweet_list)

    #extract likes and add to result
    user_likes = client.get_liked_tweets(id = id_, max_results = max_results).data
    
    
    for tweet in user_likes:
        tweet_list = []
        tweet_list.append(client.get_user(id = id_).data.id)
        tweet_list.append(client.get_user(id = id_).data.name)
        tweet_list.append(tweet.text)
        result.append(tweet_list)
    
    return result
```
