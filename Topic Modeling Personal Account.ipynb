```python
def remove_tags(tweet):
    char_list = tweet.split(' ')
    new_list = []
    for char in char_list:
        if '@' in char:
            continue
        if 'RT' in char:
            continue
        if 'https' in char:
            continue
        if '&' in char:
            continue
        else:
            new_list.append(char)

    return ' '.join(new_list)
```

```python
def remove_digits(tweet):
    char_list = tweet.split(' ')
    new_lst = []
    for char in char_list:
        if '1' in char:
            continue
        if '2' in char:
            continue
        if '3' in char:
            continue
        if '4' in char:
            continue
        if '5' in char:
            continue
        if '6' in char:
            continue
        if '7' in char:
            continue
        if '8' in char:
            continue
        if '9' in char:
            continue
        else:
            new_lst.append(char)
            
    return ' '.join(new_lst)
```

```python
df.Tweets = df.Tweets.apply(remove_digits)
```

```python
def strip_characters(tweet):
    char_list = tweet.split(' ')
    new_list = []
    for char in char_list:
        char = char.strip(':')
        char = char.strip('.')
        char = char.strip('!')
        char = char.strip('"')
        char = char.strip('?')
        char = char.strip(',')
        
        new_list.append(char)
        
    return " ".join(new_list)
```

```python
df.Tweets = df.Tweets.apply(strip_characters)
```

```python
# Function to display the top n terms in each topic. Pulled from exercises
def display_topics_lsa(model, feature_names, no_top_words, topic_names = None): 
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix + 1)
        else:
            print("\nTopic: ", topic_names[ix])
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
    print("\n")
    return model, feature_names, no_top_words
```

```python
display_topics_lsa(lsa, vec.get_feature_names(), 10, topic_names = ['economics', 'humanity', 'philosophy', 'morality'])
```
